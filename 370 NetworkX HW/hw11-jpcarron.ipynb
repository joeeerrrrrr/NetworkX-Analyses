### d.)

```python
#Calculate degree centrality, betweenness centrality, and closeness centrality of all nodes in the network

# TODO degree distrib., non-normalized
print ('Degree:', karate.degree())

# TODO degree distrib., normalized
print ('Degree centrality (normalized):', nx.degree_centrality(karate))
```

```python
# TODO find betweenness centrality (both normalized and non)
print ('Betweenness centrality (normalized):', nx.betweenness_centrality(karate, normalized=True))
print ('Betweenness centrality (unnormalized):', nx.betweenness_centrality(karate, normalized=False))
```

```python
# TODO find closeness centrality
print ('Closeness centrality:', nx.closeness_centrality(karate))
```

```python
#put the above cells into dataframe
df = pd.DataFrame(nx.degree_centrality(karate), index=['Degree Centrality'])
df2 = pd.DataFrame(nx.betweenness_centrality(karate), index=['Between Centrality'])
df3 = pd.DataFrame(nx.closeness_centrality(karate), index=['Closeness Centrality'])

tempdf = df.append(df2)
df = tempdf.append(df3)
df = df.T
df.head()
```

### d-1)

```python
resultdegree = df.sort(columns=['Degree Centrality'], ascending=False)
resultdegree
print ('Most important people in the network based on Degree Centrality are:','\n',resultdegree.ix[:0,'Degree Centrality'])
#print (df.ix['Degree Centrality'].max())
```

### d-2)

```python
resultdegree.ix[:0,:]
```

Node 33 has higher degree centrality but lower between centrality. This is intuitive because a node with high degree centrality (many links) won't have as many shortest path links between two other nodes since it is the node with the highest linkage. It is more a main central node than a node that is a bridge.

### e-2)

```python
# Visualize the community detection result using force-directed layout

# TODO: create spring layout, and save it as "pos". Use nx.spring_layout(g).
pos = nx.spring_layout(karate)

# Plot the nodes of each community using different colors.
n_communities = len(set(partition.values()))
palette = sns.color_palette('hls', n_communities)
fig, ax = plt.subplots(figsize=(12,10))
for i, community_id in enumerate(set(partition.values())):
    nodes_in_community = [nodes for nodes in partition.keys() 
                          if partition[nodes] == community_id]
    node_sizes = df.ix[nodes_in_community, 'Degree Centrality'] * 1000
    nx.draw_networkx_nodes(karate, pos, nodes_in_community, 
                           node_size = node_sizes,
                           node_color = palette[i], 
                           ax=ax)

'''Lesson: To draw edges and labels, use 
           nx.draw_networkx_edges(g) and
           nx.draw_networkx_labels(g).'''

# TODO: draw network edges and labels
nx.draw_networkx_edges(karate, pos, alpha=0.5, ax=ax)
nx.draw_networkx_labels(karate, pos, font_size=12, font_color='grey', ax=ax);

# TODO: add a title "Community Detection and Force-directed Layout" using ax.set_title()
ax.set_title('Community Detection and Force-directed Layout');
```

```python
X = nx.adjacency_matrix(g).todense()
X
```

### f-2)

```python
import sklearn.cluster as skc
import sklearn.decomposition as skd

kmeans_model = skc.KMeans(n_clusters=2).fit(X)
```

### f-3)

```python
# # PCA for dimensionality reduction
pca = skd.PCA(n_components=2)
traits_pca = pca.fit_transform(X)

# K-means
centroids = kmeans_model.cluster_centers_
centroids_pca = pca.transform(centroids)
cluster_labels = kmeans_model.labels_

# Plot the results on 2-D scatterplot with cluster centroids
plt.scatter(traits_pca[cluster_labels==0,0], 
            traits_pca[cluster_labels==0,1],
            s=100, color='b')
plt.scatter(traits_pca[cluster_labels==1,0], 
            traits_pca[cluster_labels==1,1],
            s=100, color='g')
plt.scatter(centroids_pca[:,0], centroids_pca[:,1], marker='+', s=200,
            linewidths=2, color='r')
plt.title('k-means (k=2) displayed with PCA', fontsize=15);
```
